```python
#PYTHON (Set, Dictionary)
#PYTHON SETS
thisset = {"apple", "banana", "cherry", "apple"}
print(thisset)

#the value True and 1 are considered the same value in sets
thatset = {"man", 1, True}
print(thatset)

#the values False and 0 arfe also considered the same in sets
set1 = {"man", 0, False}
print(set1)
print(len(set1))
print(type(thisset))

#Since set items are not indexed, you can only access items by using for loop ans in kryword
print(thisset)

for x in thisset:
    print(x)

print("banana" in thisset)
print("banana" not in thisset)
#Add items

thisset.add("orange")
print(thisset)

#Add a set to a set
tropical = {"pineapple", "mango", "papaya"}
thisset.update(tropical)
print(thisset)
#The update does not necessarily have to be set. it can be any iterable object
mylist = ["kiwi", "orange"]
thisset.update(mylist)
print(thisset)

#Remove Items 
thisset.remove("banana")
print(thisset)
thisset.discard("mango")
print(thisset)
thisset.clear()
print(thisset)
del thisset

#Looping items in sets 
thisset = {"apple", "banana", "cherry"}
for x in thisset:
    print(x)
    
#Join sets
#the union() method
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"John", "Elena"}
set4 = {"apple", "banana", "cherry"}
seta = set1.union(set2)
print(seta)
myset = set1.union(set2, set3, set4)
print(myset)

#the intersection() method: returns a new set that contains items that only appears in both sets
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set3 = set1.intersection(set2)
print(set3)
set1.intersection_update(set2)
print(set1)
#the difference() method:returns new set with items that are from the first set that is not in the other
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set3 = set1.difference(set2)
print(set3)
set1.difference_update(set2)
print(set1)

#the symmetric_difference() methods: keeps only elements that are not present in both sets 
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set3 = set1.symmetric_difference(set2)
print(set3)
set1.symmetric_difference_update(set2)
print(set1)

#frozenset 
x = frozenset({"apple", "banana", "cherry"})
print(x)
print(type(x))
```
