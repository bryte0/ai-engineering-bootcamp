```python
#PYTHON (Data Types)

#Data Types
x = "Hello World"
print(x)
print(type(x))
x1 = 20
print(x1)
print(type(x1))
x2 = 20.5
print(x2)
print(type(x2))
x3 = 1j
print(x3)
print(type(x3))
x4 = ["apple", "banana", "cherry"]
print(x4)
print(type(x4))
x5 = ("apple", "banana", "cherry")
print(x5)
print(type(x5))
x6 = range(6)
print(x6)
print(type(x6))
x7 = {"name" : "John", "age" : 36}
print(x7)
print(type(x7))
x8 = {"apple", "banana", "cherry"}
print(x8)
print(type(x8))
x9 = frozenset({"apple", "banana", "cherry"})
print(x9)
print(type(x9))
x10 = True
print(x10)
print(type(x10))
x11 = b"Hello"
print(x11)
print(type(x11))
x12 = bytearray(5)
print(x12)
print(type(x12))
x13 = memoryview(bytes(5))
print(x13)
print(type(x13))
x14 = None
print(x14)
print(type(x14))
```

```python
#PYTHON STRINGS 
# 'hello' is the same as "hello"

#assigning a string to variable
a = "Hello"
print(a)

#Multiline line strings can be assigned too variable by using three quotes
b = '''kwame
is a
rich guy.'''
print(b)

#strings are arrays 
a = "Hello, World"
print(a[1])

#Looping through a string
for x in "banana":
    print(x)

#String length
print(len(a))

#Check string
txt = "The best things in life are free"
print("free" in txt)
if "free" in txt:
    print("Yes, 'free' is present")

#Check if NOT
print("expensive" not in txt)
if "expensive" not in txt:
    print("No, 'expensive' is NOT present")

#Python - Slicing Strings 
b = "Hello, World"
print(b[2:5])
print(b[:5])
print(b[2:])

#Upper case 
print(b.upper())

#Lower case 
print(b.lower())

#String Concatenation
a = "Hello"
b = "World"
c = a + " " + b
print(c)

#Format strings 
age = 36
txt = f"My name is John, I am {age}"
print(txt)

price = 59
txt = f"The price is {20 * 59:.2f} dollars"
print(txt)

#Escape Characters
txt = "We are the so-called \"Vikings\" from the north"
print(txt)

#String Methods 
a = "Hello"
print(a.capitalize())
print(a.casefold())
print(a.encode())
```

```python
#Mini Project: Student Profile Card
student_name = "Bright Appiah Coffie"
age = 27
country = "Ghana"
learning = "Python"
current_language = "R"
goal = "AI Engineering"
weeks_completed = 1
total_weeks = 52
gpa = 3.85
progress_pct = (weeks_completed / total_weeks) * 100
weeks_remaining = total_weeks - weeks_completed

#Output
print("=" * 45)
print("        STUDENT PROFILE CARD")
print("=" * 45)
print(f" Name     : {student_name}")
print(f" Age      : {age}")
print(f" Country  : {country}")
print(f" From     : {current_language}")
print(f" Learning : {learning}")
print(f" Goal     : {goal}")
print("-" * 45)
print(f" Progress : {progress_pct:.1f}% complete")
print(f" Remaining: {weeks_remaining} weeks to go")
print(f" GPA      : {gpa:.2f}")
print("=" * 45);
```
