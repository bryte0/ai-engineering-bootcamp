```python
#PYTHON (Syntax, Variables)

#PYTHON SYNTAX
#Python indentation: spaces at the beginning of a code line
if 5 > 2:
    print("Five is greater than two!")
```

```python
#PYTHON VARIABLES
#Variables: created when you assign a value to it
x = 5
y = "hello, world!"
print(f"{y} is {x}") 

#Creating a Variable
x = 5
y = "John"
print(x)
print(y)
x = 4
x = "Sally"
print(x)

#many values to many variables
x, y, z = "Orange", "Banana", "Cherry"
print(x)
print(y)
print(z)

#one value to multiple variables
x = y = z = "Orange"
print(x)
print(y)
print(z)

#Output Variables
#the plus(+) sign can be used to print multiple variables
x = "Python"
y = "is"
z = "awesome"
print(x)
print(y)
print(z)
print(x, y, z)
print(x + y + z)

#Global variables
x = "Awesome"

def myfunc():
    print("Python is " + x)

myfunc()

#The global keyword
x = "awesome"

def myfunc():
     global x
     x = "fantastic"

myfunc()

print("Python is " + x)
```
